```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
medical_df = pd.read_csv(r"C:\Users\Antho\Downloads\data (6).csv")
```

```python
def create_distribution(column_name):
    feature = medical_df[column_name]
    mean = feature.mean()
    median = feature.median()
    std = feature.std()
    std_plus_one = mean + std
    std_minus_one = mean - std
    fig, ax = plt.subplots(figsize = (8, 4))
    sns.histplot(data = feature, stat = "probability", kde = True)
    ax.axvline(mean, color = "red", label = f"Mean = {mean:.2f}")
    ax.axvline(median, color = "green", label = f"Median = {median:.2f}")
    ax.axvline(std_plus_one, color = "black", label = f"+1 std = {std_plus_one:.2f}")
    ax.axvline(std_minus_one, color = "black", label = f"-1 std = {std_minus_one:.2f}")
    ax.axvspan(std_plus_one, std_minus_one, color = "yellow", zorder = 0)
    ax.set_title(f"{column_name}")
    ax.legend()
    plt.show()
```

```python
from scipy import stats
def kurtosis(column_name):
    kurt = stats.kurtosis(medical_df[column_name], fisher = False)
    return (f"{kurt:.2f}")
```

```python
create_distribution("VitD_levels")
```

```python
kurtosis("VitD_levels")
```

### The distribution is continuous. The distribution has a slight negative skew. The graph displays a mesokurtic kurtosis.

```python
kurtosis("Doc_visits")
```
